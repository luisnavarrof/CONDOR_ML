```python
# Ruta al archivo combined_stats.csv
stats_path = r'C:\Users\Froxo\OneDrive - Universidad Tcnica Federico Santa Mara\ML_Predictions\Predicts_All\ResNet50_stats_gamma.csv'

# Leer el archivo CSV
data = pd.read_csv(stats_path)

# Obtener los ngulos nicos en los datos
unique_angles = sorted(data['True Angle'].unique())

# Inicializar listas para los valores promediados y las barras de error
true_angles = []
predicted_means = []
error_bars = []

# Calcular valores promedio y desviacin estndar para cada ngulo
for angle in unique_angles:
    # Filtrar las predicciones para el ngulo actual
    predictions = data[data['True Angle'] == angle]['Predicted Angle']
    
    # Calcular el promedio y la desviacin estndar
    mean_prediction = predictions.mean()
    std_deviation = predictions.std()
    
    # Agregar los valores calculados a las listas
    true_angles.append(angle)
    predicted_means.append(mean_prediction)
    error_bars.append(std_deviation)  # Aqu se asigna la desviacin estndar

# Crear figura combinada
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Subgrfico 1: Dispersin de ngulos verdaderos vs predichos
axs[0].scatter(data['True Angle'], data['Predicted Angle'], alpha=0.03, color='orange', label='Predictions')
axs[0].plot([0, max(unique_angles)], [0, max(unique_angles)], 'k--', label='Reference (y = x)')
axs[0].set_xlabel('True Angle ()')
axs[0].set_ylabel('Predicted Angle ()')
axs[0].set_title('True Angle vs Predicted Angle')
axs[0].legend()
axs[0].grid(True)

# Subgrfico 2: Promedios con barras de error (+/- 1 desviacin estndar)
axs[1].errorbar(true_angles, predicted_means, yerr=error_bars, fmt='o', color='blue', 
                ecolor='red', capsize=5, elinewidth=1, markeredgewidth=1, 
                label='Predicted Mean  1 STD')
axs[1].plot(unique_angles, unique_angles, linestyle='--', color='gray', label='Ideal Prediction (y = x)')
axs[1].set_title('Mean Predicted Angles with Error Bars (+/- 1 STD)')
axs[1].set_xlabel('True Angle ()')
axs[1].set_ylabel('Predicted Angle ()')
axs[1].grid(linestyle='--', alpha=0.7)
axs[1].legend()

# Ttulo general para la figura
fig.suptitle('ResNet50 Model Predictions\nGamma Ray (Photon), 800 GeV, 20 Epochs, Binned data', fontsize=14)

# Ajustar diseo y reducir espacio entre el ttulo y los grficos
plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta al archivo CSV
csv_path = r"C:\Users\Froxo\OneDrive - Universidad Tcnica Federico Santa Mara\ML_Predictions\LSTM_Predicts_Energy_2E2\LSTM_Predicts_Energy_2E2.csv"

# Cargar el CSV
data = pd.read_csv(csv_path)


# Verificar columnas
print("Columnas disponibles:", data.columns.tolist())
if "Real_Particle" not in data.columns or "Predicted_Particle" not in data.columns:
    print("El archivo no tiene las columnas necesarias: 'Real_Particle' o 'Predicted_Particle'.")
    exit()

# Manejar valores faltantes en Predicted_Particle
data["Predicted_Particle"] = pd.to_numeric(data["Predicted_Particle"], errors="coerce")  # Convertir a numrico (NaN si falla)
data = data.dropna(subset=["Predicted_Particle"])  # Eliminar filas con NaN
data["Predicted_Particle"] = data["Predicted_Particle"].astype(int)  # Convertir a entero despus de limpiar

# Grfico 1: Matriz de confusin visual (Real vs Prediccin)
plt.figure(figsize=(6, 4))
sns.heatmap(
    pd.crosstab(data["Real_Particle"], data["Predicted_Particle"]),
    annot=True,
    fmt="d",
    cmap="Blues",
    cbar=True
)
plt.title("Matriz de Confusin: Real vs Prediccin")
plt.xlabel("Partcula Predicha")
plt.ylabel("Partcula Real")
plt.tight_layout()
plt.show()

# Contar predicciones correctas
correct_predictions = (data["Real_Particle"] == data["Predicted_Particle"]).sum()
total_predictions = len(data)
accuracy = correct_predictions / total_predictions * 100

# Mostrar cantidad de predicciones correctas
print(f"Predicciones correctas: {correct_predictions}/{total_predictions} ({accuracy:.2f}%)")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Parmetros
energy = "2E2"
predicts_dir = f"C:\\Users\\Froxo\\OneDrive - Universidad Tcnica Federico Santa Mara\\ML_Predictions\\LSTM_Predicts_ProtonPhoton_{energy}"
loss_file_name = os.path.join(predicts_dir, f'LSTM_Loss_ProtonPhoton_{energy}.csv')
stats_file_name = os.path.join(predicts_dir, f'LSTM_Predicts_ProtonPhoton_{energy}.csv')

# Verificar existencia de archivos
if not os.path.exists(loss_file_name):
    print(f"No se encontr el archivo de prdida: {loss_file_name}")
    exit()

if not os.path.exists(stats_file_name):
    print(f"No se encontr el archivo de estadsticas: {stats_file_name}")
    exit()

# Leer datos
loss_data = pd.read_csv(loss_file_name)
stats_data = pd.read_csv(stats_file_name)

# Graficar precisin a travs de las epochs
plt.figure(figsize=(12, 6))
plt.plot(loss_data['epoch'], loss_data['accuracy'], label='Entrenamiento', marker='o')
plt.plot(loss_data['epoch'], loss_data['val_accuracy'], label='Validacin', marker='o')
plt.title('Precisin a travs de las Epochs')
plt.xlabel('Epoch')
plt.ylabel('Precisin')
plt.legend()
plt.grid()

# Guardar el grfico de precisin
accuracy_plot_path = os.path.join(predicts_dir, f'LSTM_Accuracy_Plot_ProtonPhoton_{energy}.png')
plt.savefig(accuracy_plot_path, dpi=300)
plt.show()
print(f"Grfico de precisin guardado en: {accuracy_plot_path}")

# Matriz de confusin
true_labels = stats_data['True Label']
predicted_labels = stats_data['Predicted Label']

# Crear matriz de confusin
conf_matrix = confusion_matrix(true_labels, predicted_labels)
disp = ConfusionMatrixDisplay(conf_matrix, display_labels=['Protones (0)', 'Fotones (1)'])

# Graficar la matriz de confusin
disp.plot(cmap='Blues', colorbar=True)
plt.title('Matriz de Confusin')
conf_matrix_path = os.path.join(predicts_dir, f'LSTM_Confusion_Matrix_ProtonPhoton_{energy}.png')
plt.savefig(conf_matrix_path, dpi=300)
plt.show()
print(f"Matriz de confusin guardada en: {conf_matrix_path}")

# Grfico mejorado: Valores reales vs predicciones
plt.figure(figsize=(10, 6))
colors = ['blue' if label == 0 else 'red' for label in true_labels]  # Azul para protones, rojo para fotones

plt.scatter(range(len(true_labels)), true_labels, label='Etiqueta Real', alpha=0.6, c=colors, marker='o')
plt.scatter(range(len(predicted_labels)), predicted_labels, label='Clasificacin del Modelo', alpha=0.6, c='green', marker='x')
plt.title('Valores Reales vs Clasificaciones del Modelo (Mejorado)')
plt.xlabel('ndice')
plt.ylabel('Etiqueta')
plt.legend()
plt.grid()

# Guardar el grfico de valores reales vs. predicciones
true_vs_predicted_path = os.path.join(predicts_dir, f'LSTM_True_vs_Predicted_ProtonPhoton_{energy}_Improved.png')
plt.savefig(true_vs_predicted_path, dpi=300)
plt.show()
print(f"Grfico mejorado de valores reales vs clasificaciones guardado en: {true_vs_predicted_path}")


```
