```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Área en metros cuadrados
area = 122 * 113

# Directorios de los archivos
directories = {
    'Oxygen': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_O\binned_data",
    'Carbon': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_C\binned_data",
    'Iron': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_Fe\binned_data",
    'Helium-4': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_He4\binned_data",
    'Proton': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_Proton\binned_data",
}

# Función para calcular la densidad promedio de partículas
def calculate_average_particle_density(directory, area):
    total_particle_count = 0
    csv_files = glob.glob(os.path.join(directory, "*.csv"))
    total_files = len(csv_files)

    if total_files == 0:
        return 0, total_files

    for filepath in csv_files:
        try:
            data = pd.read_csv(filepath)
            if 'particle_count' in data.columns:
                total_particle_count += data['particle_count'].sum()
            else:
                print(f"Warning: 'particle_count' column missing in {filepath}")
        except Exception as e:
            print(f"Error reading {filepath}: {e}")

    average_particle_count = total_particle_count / total_files
    particle_density = average_particle_count / area
    return particle_density, total_files

# Diccionario para densidades por ángulo
angle_densities = {label: [] for label in directories.keys()}

# Calcular densidades para cada directorio y carpeta run
for label, base_dir in directories.items():
    for run_number in range(1, 32):
        run_dir = os.path.join(base_dir, f"run_{run_number}")
        if os.path.isdir(run_dir):
            angle = (run_number - 1) * 2
            density, file_count = calculate_average_particle_density(run_dir, area)
            angle_densities[label].append((angle, density))

# Crear carpeta para resultados
output_dir = r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\Density_Data\Cosmic_Rays"
os.makedirs(output_dir, exist_ok=True)

# Guardar datos y graficar
plt.figure(figsize=(12, 8))
for label, densities in angle_densities.items():
    if len(densities) == 0:
        print(f"No data for {label}. Skipping.")
        continue

    angles, density_values = zip(*densities)
    density_data = pd.DataFrame({'angle': angles, 'density': density_values})
    output_path = os.path.join(output_dir, f"density_{label.replace(' ', '_')}.csv")
    density_data.to_csv(output_path, index=False)
    print(f"Saved density data for {label} to {output_path}")

    # Graficar densidades
    sns.lineplot(x=angles, y=density_values, label=label)

# Configurar gráfico
plt.title("Particle Density by Angle and Energy")
plt.xlabel("Angle (degrees)")
plt.ylabel("Particle Density (particles/m²)")
plt.legend(title="Energy")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Área en metros cuadrados
area = 122 * 113

# Directorios de los archivos
directories = {
    'Oxigeno': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_O\binned_data",
    'Carbono': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_C\binned_data",
    'Hierro': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_Fe\binned_data",
    'He4': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_He4\binned_data",
    'Proton': r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\CONDOR_Angle_2\CONDOR_2E2_Proton\binned_data",
}

# Función para calcular la densidad promedio de partículas en un directorio
# Función para calcular la densidad promedio de partículas en un directorio
def calculate_average_particle_density(directory, area):
    total_particle_count = 0
    total_files = 1000  # Usamos 1000 archivos como el total esperado
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            filepath = os.path.join(directory, filename)
            data = pd.read_csv(filepath)
            total_particle_count += data['particle_count'].sum()
    
    # Calculamos la densidad utilizando el total de archivos esperado (1000)
    average_particle_count = total_particle_count / total_files
    particle_density = average_particle_count / area
    return particle_density

# Inicializar un diccionario para almacenar densidades por ángulo
angle_densities = {label: [] for label in directories.keys()}

# Iterar sobre cada directorio y cada carpeta "run"
for label, base_dir in directories.items():
    for run_number in range(1, 32):
        run_dir = os.path.join(base_dir, f"run_{run_number}")
        if os.path.isdir(run_dir):
            angle = (run_number - 1) * 2  # Calcular el ángulo
            density = calculate_average_particle_density(run_dir, area)
            angle_densities[label].append((angle, density))

# Ordenar densidades por ángulo
for label in angle_densities:
    angle_densities[label].sort()

# Crear carpeta para guardar resultados si no existe
output_dir = r"C:\Users\Froxo\OneDrive - Universidad Técnica Federico Santa María\Density_Data\Cosmic_Rays_2try"
os.makedirs(output_dir, exist_ok=True)

# Guardar los datos de densidad sin suavizar en un archivo CSV
for label, densities in angle_densities.items():
    if len(densities) == 0:  # Comprobar si la lista está vacía
        print(f"No data found for {label}. Skipping.")
        continue
    
    # Extraer ángulos y densidades
    angles, density_values = zip(*densities)
    
    # Crear DataFrame para guardar los datos
    density_data = pd.DataFrame({
        'angle': angles,
        'density': density_values
    })
    
    # Guardar el archivo CSV
    output_path = os.path.join(output_dir, f"density_{label.replace(' ', '_')}.csv")
    density_data.to_csv(output_path, index=False)
    print(f"Saved density data for {label} to {output_path}")



```
